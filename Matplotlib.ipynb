```python
import numpy as np
```

```python
plt.plot(x, np.cos(x))
# cos그래프를 그린다
```

```python
plt.subplot(2,1,1)
# 한 화면안에 두개의 그래프를 표현한다. 
# (2,1,1) 2개를 표현할건데 1줄짜리로 1번째다
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
# (2,1,2) 2개를 표현할건데 1줄짜리로 2번째다.
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x-0), color='blue')
# 선의 색깔을 color 로 지정할수있다.
plt.plot(x,np.sin(x-1), color='g')
# 선의 위치를 바꿀수있다. g= green (짧은색상코드) , b=blue, k =black (둘다 b가 시작이므로)
plt.plot(x,np.sin(x-2), color='0.5')
# color 에 숫자를 입력하면 1이 밝색, 0이 어두운색이다. 음영.
plt.plot(x,np.sin(x-3), color='#FFDD44')
# color '#' 을 사용하면  16진수로 색상의 코드(RGB)를 지정할수있다.
plt.plot(x,np.sin(x-4), color=(1.0,0.2,0.3))
# 튜플로써 10진수로도 표현이 가능하다. (0~1사이)(RGB)
plt.plot(x,np.sin(x-5), color='chartreuse')
# 컬러의 이름을 지정해줄수도 있다.
```

```python
plt.plot(x,x+0,linestyle='solid')
#  linestyle : 선의 형태를 지정할수있다
plt.plot(x,x+1,linestyle='dashed')
# dashed : dash로 이루어진 선.
plt.plot(x,x+2,linestyle='dashdot')
# dash 와 dot 이 번갈아가면서 나오는 선
plt.plot(x,x+3,linestyle='dotted')
# dotted : dot 으로 이루어진 선


# 위와 같은데 표현이 다르다.

plt.plot(x,x+4,linestyle='-')
#  - : solid
plt.plot(x,x+5,linestyle='--')
# -- : dashed
plt.plot(x,x+6,linestyle='-.')
# -. :  dashdot
plt.plot(x,x+7,linestyle=':')
# :  : dotted
```

```python
plt.plot(x,x+0,'-g')
# solid 라인으로 g(green) 선이 그어진다
plt.plot(x,x+1,'--c')
# dash 선으로 시안색
plt.plot(x,x+2,'-.k')
# 대시+점 으로 검은색
plt.plot(x,x+3,':r')
# dot 선응로 빨간색
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)
# xlim : x의 한계를 걸어준다.
plt.ylim(1.2,-1.2)
# ylim: y의 한계를 걸어준다.

#음수의값과 양수의 값의 순서를 바꾸면 그래프가 뒤집힌다. ★반대로도 작동을 한다
```

```python
plt.plot(x,np.sin(x),'-g', label='sin(x)')
# label : 텍스트를 출력하는 것이다.
```

```python
plt.plot(x, np.sin(x),'-g', label='sin(x)')
plt.plot(x, np.cos(x),':b', label='cos(x)')
plt.axis('equal')
# 두개의 플롯의 축을 지정한다.
plt.legend()
#   라벨정보가 없으면 작동을 하지않는다, 우측상단에 정보가 표시된다.
```

```python
x = np.linspace(0,10,30)
plt.plot(x,np.sin(x),'o',color='black')
# 알파벳 o : 산점도 : 선 대신에 굵은 점이 찍히는것 (한개 한개의 정보가 중요할때 사용)
# 산점도에 찍히는 점들을 marker 라고 칭한다.
```

```python
rng = np.random.RandomState(0)
# .random.RandomState : 랜덤값을 만들어낸다.()안에 수치를 입력하면 시드값이 정해진다.
for marker in ['o', ',',',','x','+','v','^','<','>','s','d']: # marker 들이 하나씩 마커에 대입되서 반복된다.
    plt.plot(rng.rand(5), rng.rand(5), marker, 
             label='marker={0}'.format(marker))
    # 5안에서 값이 잡힌다.
    # 사용된 marker 에 따라서 marker 에 대한 label 을 지정해준다.
    plt.legend()
```

```python
y = np.sin(x)
plt.plot(x,y,'-ok')
# -o 산점도를 선으로 긋는다, -ok : 검은색으로 바꾼다.
```

```python
plt.plot(x,y,'-p',color='gray',
        markersize=15,linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth='2')
# p : pentagon 오각형
```

```python
plt.scatter(x,y)
# scatter : 원으로 찍힌다.
```

```python
rng = np.random.RandomState(0)
# randomstate () 안의 숫자를 지정함으로써 시드값을 고정할수있다.
x = rng.randn(100)
# randn 의 끝의 n의 갯수를 뜻한다 ex) randn(100) 은 100개의 랜덤한 값이 찍힌다.
# 범위를 별다른 지정 하지 않으면 -3~ +5 사이의 값으로 정해진다.
y = rng.randn(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)
# rand 는 양수의 범위에서만 생성이 된다, randn 은 음수의 범위에서 생성이 된다.

plt.scatter(x,y, c=color, s=sizes, alpha=0.3, cmap='viridis')
# color 를 c로 줄여서 사용할수도있다.
# size 를 s로 줄여서 사용할수도있다.
# alpha 값으로 색상의 투명도를 설정할수있다.
plt.colorbar()
#색상표 를 불러올수있다.
```

```python
from sklearn.datasets import load_iris
# sklearn.datasets  에도 iris 샘플데이터가 있으며, 그것을 불러온다.
iris = load_iris()
# NumPy 배열로 저장되어있다.
features = iris.data.T
# iris 데이터중 T값만 가져온다. T: Target 값.
```

```python
plt.scatter(features[0],features[1], alpha=0.2 , s=features[3]*100, cmap ='viridis',
           c=iris.target)
# 0번째 feature : 꽃잎의 길이, 1번째 feature : 꽃잎의 폭 3번째 : 꽃받침의 길이
plt.xlabel(iris.feature_names[0])
# x에 iris feature name 의 0번째 값을 가져온다.
plt.ylabel(iris.feature_names[1])
# 1번째 iris feature name의 값을 y에 적용한다
```

```python
#errorbar : 오차를 표현할수있다.
x = np.linspace(0,10,50)
dy = 0.8
# dy : 오차값
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x,y, yerr =dy, fmt='.k')
# yerr : y의 error 아래 그림의 각포인트에 있는 선이 나올수있는 에러의 범위이다.
# fmt : 형태를 지정한다.
```

```python
plt.style.use('seaborn-whitegrid')
# plot의 style을 지정할수있다.
plt.errorbar(x,y,yerr=dy,fmt='o', color='black',
            ecolor='lightgray',elinewidth =3)
# ecolor : error 의 색깔
# elinewidth : error line 의 폭
```
